```python
#importing the libraries
from bs4 import BeautifulSoup as bs
import requests
import time
```

```python
#taking input from user
num=int(input('No. of pages:'))
```

```python
#collecting the url of different pages
headers={'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36 Edg/91.0.864.59'}
pages=list(range(1,num))
a=[]
for i in pages:
    url=requests.get('https://www.amazon.in/product-reviews/8192910911/ref=cm_cr_arp_d_viewopt_sr?ie=UTF8&filterByStar=all_stars&reviewerType=all_reviews&pageNumber={}#reviews-filter-bar'.format(i),headers=headers)
    time.sleep(3)
    a.append(url)
    
```

```python
#scrapping the url collected using bs4
soups=[]
for i in a:
    soup=bs(i.content,'html.parser')
    print(soup.prettify())
    soups.append(soup)
```

```python
#finding all customer name 
name=[]
for i in soups:
    na=i.find_all('span',{'class':'a-profile-name'})
    name.append(na)
    

```

```python
name
```

```python
cust_name=[]
for  i in name:
    for j in i:
        cust_name.append(j.get_text())
           
```

```python
cust_name
```

```python
#removing duplicated names
cust_name1=[]
for i in cust_name:
    if i not in cust_name1:
        cust_name1.append(i)
```

```python
cust_name1
```

```python
#finding all reviews
review=[]
for i in soups:
    reviews=i.find_all('span',{'data-hook':'review-body'})
    review.append(reviews)
```

```python
review
```

```python
cust_review=[]
for i in review:
    for j in i:
        cust_review.append(j.get_text())
cust_review
```

```python
#removing duplicated review
cust_review1=[]
for i in cust_review:
    if i not in cust_review1:
        cust_review1.append(i)
```

```python
cust_review1
```

```python
#cleaning a bit
cust_review1= [a.strip('\n') for a in cust_review1]
#review =[a.rstrip('\n') for a in review]
```
